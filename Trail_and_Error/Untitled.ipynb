```python
def clean_json(pr):
    del_keys = []
    for key in pr:
        if key[0] == "_":
            del_keys.append(key)
        else:
            val = pr[key]
            if issubclass(type(val), dict):
                clean_json(val)

    for key in del_keys:
        del pr[key] 
```
